# Task 02: Exploratory Data Analysis (EDA) on the Titanic Dataset

In this notebook, we will perform data cleaning and EDA on the Titanic dataset from Kaggle.

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
df = pd.read_csv(url)

# Display first few rows
df.head()
```

## Step 1: Basic Information About the Dataset

```python
# Check shape and basic info
print("Shape of dataset:", df.shape)
df.info()
```

## Step 2: Handling Missing Values

```python
# Check missing values
df.isnull().sum()

# Fill missing Age with median and Embarked with mode
df['Age'].fillna(df['Age'].median(), inplace=True)
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
df.drop(columns=['Cabin'], inplace=True)  # Too many missing values
```

## Step 3: Exploratory Data Analysis (EDA)

```python
# Plot survival count
sns.countplot(x='Survived', data=df)
plt.title('Survival Count')
plt.show()
```

```python
# Survival by gender
sns.countplot(x='Survived', hue='Sex', data=df)
plt.title('Survival by Gender')
plt.show()
```

```python
# Age distribution
sns.histplot(df['Age'], bins=30, kde=True)
plt.title('Age Distribution')
plt.show()
```

```python
# Survival by class
sns.countplot(x='Survived', hue='Pclass', data=df)
plt.title('Survival by Passenger Class')
plt.show()
```
