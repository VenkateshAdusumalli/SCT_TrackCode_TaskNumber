# Task 01: Visualizing India's Population Distribution by Age (2022)

This notebook creates a bar chart to show the distribution of India's population across age groups using simple Python code.

```python
import matplotlib.pyplot as plt

# Age groups and their population
age_groups = ['0-20 Years', '21-64 Years', '65+ Years']
population_millions = [512, 807, 98]

# Create the bar chart
plt.bar(age_groups, population_millions, color=['yellow', 'blue', 'magenta'])

# Add titles and labels
plt.title("India's Population Distribution by Age (2022)")
plt.xlabel("Age Groups")
plt.ylabel("Population (in Millions)")

# Show the plot
plt.show()
```
